## Breaking Bad: A Brief Description

*Breaking Bad* is an American television series created by Vince Gilligan. It aired from January 20, 2008, to September 29, 2013. The show follows Walter White (played by Bryan Cranston), a high school chemistry teacher turned methamphetamine manufacturer, and his former student and business partner, Jesse Pinkman (played by Aaron Paul).

As the series progresses, Walter White's transformation from a mild-mannered teacher to a ruthless drug lord is explored, along with the consequences of his actions on his family and the criminal underworld.

*Breaking Bad* received widespread critical acclaim for its writing, acting, and character development. It won numerous awards, including 16 Primetime Emmy Awards.


## Ratings

*Breaking Bad premiered on the same night as both the NFC and AFC Championships in the 2008 NFL playoffs, an intentional decision by AMC hoping to capture the adult male viewership immediately following the planned end of the NFC game. The game ran over its time slot, cutting into Breaking Bad's timeslot in most of America. As a result, the pilot had only about 1.4 million viewers. Coupled with the ongoing writers strike, the first season did not draw as large of a viewership as they expected. However, with subsequent seasons, viewership increased, avoiding the usual trend of downward viewership that most serialized shows had. Ratings further increased by the fourth season as, prior to airing, the previous seasons had been added to Netflix, boosting interest in the show. Breaking Bad is considered the first such show to have a renewed burst of interest due to the show being made available on Netflix. The second half of the final season saw record viewership, with the series finale reaching over 10.3 million viewers.*

## A graph of viewership over time and season-to-season changes in viewership

```python
import matplotlib.pyplot as plt

# Data
seasons = ["Season 1", "Season 2", "Season 3", "Season 4", "Season 5A", "Season 5B"]
viewers = [1.23, 1.30, 1.52, 1.87, 4.32, 10.28]

# Graph of Viewership Over Time
plt.figure(figsize=(10, 5))
plt.plot(seasons, viewers, marker='o', color='b', linestyle='-')
plt.title('Viewership Over Time')
plt.xlabel('Season')
plt.ylabel('Average Viewership (millions)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Graph of Episode-to-Episode Changes in Viewership
viewers_diff = [0] + [viewers[i] - viewers[i-1] for i in range(1, len(viewers))]
plt.figure(figsize=(10, 5))
plt.bar(seasons, viewers_diff, color='g')
plt.title('Season-to-Season Changes in Viewership')
plt.xlabel('Season')
plt.ylabel('Change in Viewership (millions)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Observed Changes in Viewership

***The viewership of "Breaking Bad" exhibited notable fluctuations throughout its five seasons. The average viewership increased steadily from Season 1 to Season 4, with each subsequent season surpassing the previous in terms of viewership numbers. For instance, the average viewership rose from 1.23 million in Season 1 to 1.87 million in Season 4, indicating a steady growth trend.***

***However, Season 5 of "Breaking Bad" saw a remarkable surge in viewership, particularly during its second half (Season 5B), where the average viewership peaked at 10.28 million. This increase can be attributed to various factors, including heightened anticipation for the series finale and increased word-of-mouth promotion due to the show's growing popularity.***

***Interestingly, the increase in viewership between Seasons 4 and 5A was substantial, with the average viewership more than doubling from 1.87 million to 4.32 million. This spike likely reflects the heightened tension and anticipation as the series approached its conclusion.***

***In contrast, the observed changes in viewership between Seasons 3 and 5 are particularly noteworthy. During this period, the viewership increased significantly by 8.76 million, showcasing the show's ability to captivate and retain a growing audience over time.***

**Overall, the fluctuations in viewership throughout "Breaking Bad" underscore its ability to engage viewers and maintain a strong following, ultimately solidifying its status as one of the most iconic television series of all time.**
